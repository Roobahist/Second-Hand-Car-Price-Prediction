```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import keras
from keras.models import Sequential
from keras.layers import *
from keras.models import Model
from keras import Input, layers
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error , r2_score
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA
```

## Splitting Data into 5-folds

```python
slicing_indices_for_kfold = [0,2275,4549,6823,9098,11372]
```

```python
five_fold_dict = {
    'one' : {
        'train' : [i for i in range (11372) if i not in list(np.arange(2275))] ,
        'test' : list(np.arange(2275))
    } ,
    'two' : {
        'train' : [i for i in range (11372) if i not in list(np.arange(2275,4549))] ,
        'test' : list(np.arange(2275,4549))
    } ,
    'three' : {
        'train' : [i for i in range (11372) if i not in list(np.arange(4549,6823))] ,
        'test' : list(np.arange(4549,6823))
    } ,
    'four' : {
        'train' : [i for i in range (11372) if i not in list(np.arange(6823,9098))] ,
        'test' : list(np.arange(6823,9098))
    } ,
    'five' : {
        'train' : [i for i in range (11372) if i not in list(np.arange(9098,11372))] ,
        'test' : list(np.arange(9098,11372))
    } ,
}
```

```python
for k,v in five_fold_dict.items():
    print('fold number =',k)
    print('train size =',len(v['train']))
    print('test size =',len(v['test']))
    print('--------------')
```

```python
def index_to_x_y(Dataframe , list_of_indices):
    x = Dataframe.iloc[list_of_indices,:].drop('price',axis=1)
    y = Dataframe.iloc[list_of_indices,:]['price']
    return x, y
```

# Regression

```python
reg_encoded = pd.read_csv('../input/car-price-estimation-ai-data/reg_encoded.csv').drop('Unnamed: 0',axis=1)
reg_encoded = reg_encoded.reset_index(drop=True)
reg_encoded.head()
```

```python
reg_models = []
reg_mean_absolute_errors_train = []
reg_mean_absolute_errors_test = []
reg_r2_scores_train = []
reg_r2_scores_test = []
reg_train_predictions = []
reg_y_trains = []
reg_test_predictions = []
reg_y_tests = []
for k , v in five_fold_dict.items():
    x_train , y_train = index_to_x_y(reg_encoded , v['train'])
    x_test , y_test = index_to_x_y(reg_encoded , v['test'])
    reg = LinearRegression().fit(x_train,y_train)
    reg_models.append(reg)
    y_hat_train = reg.predict(x_train)
    y_hat_test = reg.predict(x_test)
    mae_train = mean_absolute_error(y_train,y_hat_train)
    mae_test = mean_absolute_error(y_test,y_hat_test)
    r2_train = r2_score(y_train,y_hat_train)
    r2_test = r2_score(y_test,y_hat_test)
    reg_train_predictions.append(y_hat_train)
    reg_test_predictions.append(y_hat_test)
    reg_mean_absolute_errors_train.append(mae_train)
    reg_mean_absolute_errors_test.append(mae_test)
    reg_r2_scores_train.append(r2_train)
    reg_r2_scores_test.append(r2_test)
    reg_y_trains.append(y_train)
    reg_y_tests.append(y_test)
    
```

```python
reg_metrics = {
    'train_mae' : np.mean(reg_mean_absolute_errors_train)/1000000 ,
    'test_mae' : np.mean(reg_mean_absolute_errors_test)/1000000 ,
    'r2_train' : np.mean(reg_r2_scores_train) ,
    'r2_test' : np.mean(reg_r2_scores_test)
}

print('Regression')
print("mean of mae's for 5-fold on training records =","%0.2f" % (np.mean(reg_mean_absolute_errors_train)/1000000),'milion')
print("mean of mae's for 5-fold on test records =","%0.2f" % (np.mean(reg_mean_absolute_errors_test)/1000000),'milion')
print("mean of r2_score's for 5-fold on training records =",np.mean(reg_r2_scores_train))
print("mean of r2_score's for 5-fold on test records =",np.mean(reg_r2_scores_test))
```

```python
def upper_rugplot(data, height=.05, ax=None, **kwargs):
    from matplotlib.collections import LineCollection
    ax = ax or plt.gca()
    kwargs.setdefault("linewidth", 1)
    segs = np.stack((np.c_[data, data],
                     np.c_[np.ones_like(data), np.ones_like(data)-height]),
                    axis=-1)
    lc = LineCollection(segs, transform=ax.get_xaxis_transform(), **kwargs)
    ax.add_collection(lc)
```

```python
fig,ax = plt.subplots(nrows=5,ncols=2,figsize=(15,25))
fig.tight_layout(pad=5.0)
for i in range(5):
    ax[i][0].set_title('Fold '+str(i)+' Train Residuals')
    ax[i][0].set_xlabel('Error')
    ax[i][0].set_ylabel('Number')
    sns.distplot(abs(reg_y_trains[i]-reg_train_predictions[i]),ax=ax[i][0])
    upper_rugplot(abs(reg_y_trains[i]-reg_train_predictions[i]),ax=ax[i][0],color='red')

    ax[i][1].set_title('Fold '+str(i)+' Test Residuals')
    ax[i][1].set_xlabel('Error')
    ax[i][1].set_ylabel('Number')
    sns.distplot(abs(reg_y_tests[i]-reg_test_predictions[i]),ax=ax[i][1],color='teal')
    upper_rugplot(abs(reg_y_tests[i]-reg_test_predictions[i]),ax=ax[i][1],color='red')
```

```python
def make_network():
    adam = keras.optimizers.adam(learning_rate=0.001)
    network = Sequential()
    network.add(Dense(300,input_shape=(7,),activation='relu',kernel_initializer='normal'))
    network.add(Dense(550,activation='relu',kernel_initializer='normal'))
    network.add(Dense(550,activation='relu',kernel_initializer='normal'))
    network.add(Dense(550,activation='relu',kernel_initializer='normal'))
    network.add(Dense(550,activation='relu',kernel_initializer='normal'))
    network.add(Dense(550,activation='relu',kernel_initializer='normal'))
    network.add(Dense(550,activation='relu',kernel_initializer='normal'))
    network.add(Dense(550,activation='relu',kernel_initializer='normal'))
    network.add(Dense(550,activation='relu',kernel_initializer='normal'))
    network.add(Dense(550,activation='relu',kernel_initializer='normal'))
    network.add(Dense(300,activation='relu',kernel_initializer='normal'))
    network.add(Dense(1))
    network.compile(loss='mae',optimizer=adam)
    return network
```

```python
network = make_network()
network.summary()
```

```python
neu_networks_history = []
neu_mean_absolute_errors_train = []
neu_mean_absolute_errors_test = []
neu_r2_scores_train = []
neu_r2_scores_test = []
neu_train_predictions = []
neu_y_trains = []
neu_test_predictions = []
neu_y_tests = []
i = 0
for k,v in five_fold_dict.items():
    i += 1
    x_train , y_train = index_to_x_y(neu_encoded , v['train'])
    x_test , y_test = index_to_x_y(neu_encoded , v['test'])
    network = make_network()
    print(network.optimizer.learning_rate)
    checkpoint = keras.callbacks.ModelCheckpoint(filepath='../working/weights'+str(i)+'.h5',monitor='val_loss', verbose=1, save_best_only=True, mode='min')
    reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.8,patience=5, min_lr=0.00001)
    callbacks_list = [reduce_lr,checkpoint]
    network.fit(x_train,y_train,epochs=200,validation_data=(x_test,/),callbacks=callbacks_list)
    print(network.optimizer.learning_rate)
    y_hat_train = network.predict(x_train)
    y_hat_test = network.predict(x_test)
    mae_train = mean_absolute_error(y_train,y_hat_train)
    mae_test = mean_absolute_error(y_test,y_hat_test)
    r2_train = r2_score(y_train,y_hat_train)
    r2_test = r2_score(y_test,y_hat_test)
    neu_networks_history.append(network.history.history)
    neu_train_predictions.append(y_hat_train)
    neu_test_predictions.append(y_hat_test)
    neu_mean_absolute_errors_train.append(mae_train)
    neu_mean_absolute_errors_test.append(mae_test)
    neu_r2_scores_train.append(r2_train)
    neu_r2_scores_test.append(r2_test)
    neu_y_trains.append(y_train)
    neu_y_tests.append(y_test)
```

```python
neu_metrics = {
    'train_mae' : np.mean(neu_mean_absolute_errors_train)/1000000 ,
    'test_mae' : np.mean(neu_mean_absolute_errors_test)/1000000 ,
    'r2_train' : np.mean(neu_r2_scores_train) ,
    'r2_test' : np.mean(neu_r2_scores_test)
}

print('Neural Network')
print("mean of mae's for 5-fold on training records =","%0.2f" % (np.mean(neu_mean_absolute_errors_train)/1000000),'milion')
print("mean of mae's for 5-fold on test records =","%0.2f" % (np.mean(neu_mean_absolute_errors_test)/1000000),'milion')
print("mean of r2_score's for 5-fold on training records =",np.mean(neu_r2_scores_train))
print("mean of r2_score's for 5-fold on test records =",np.mean(neu_r2_scores_test))
```

```python
neu_best_test_loss = []
neu_corresponding_train_loss = []
for i in range(len(neu_networks_history)):
    min_index = 0
    for j in range(len(neu_networks_history[i]['val_loss'])):
        if neu_networks_history[i]['val_loss'][j] < neu_networks_history[i]['val_loss'][min_index]:
            min_index = j
    neu_best_test_loss.append(neu_networks_history[i]['val_loss'][j])
    neu_corresponding_train_loss.append(neu_networks_history[i]['loss'][j])
    
neu_best_performance = pd.DataFrame()
neu_best_performance['Train'] = neu_corresponding_train_loss
neu_best_performance['Test'] = neu_best_test_loss
```

```python
N = 5
fig, ax = plt.subplots(figsize=(10,7))
ind = np.arange(N)    # the x locations for the groups
width = 0.35         # the width of the bars
p1 = ax.bar(ind, neu_best_performance['Train'], width)
p2 = ax.bar(ind + width, neu_best_performance['Test'], width)

ax.set_title('Neural Network best Test Loss and corresponding Train Loss on 5-fold')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('fold1', 'fold2', 'fold3', 'fold4', 'fold5'))
ax.set_ylabel('Loss')

ax.legend((p1[0], p2[0]), ('Train', 'Test'))
ax.autoscale_view()
```
